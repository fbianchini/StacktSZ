```python
anal = {}
anal[5] = CutoutAnalysis('results_filt_H-ATLAS_W4_SN5', zbins=zbins, extras_names=extras_names, lambdas=lambdas)
anal[4] = CutoutAnalysis('results_filt_H-ATLAS_W4_SN4', zbins=zbins, extras_names=extras_names, lambdas=lambdas)
anal[3] = CutoutAnalysis('results_filt_H-ATLAS_W4_SN3/', zbins=zbins, extras_names=extras_names, lambdas=lambdas)
```

```python
fluxes = {}

for lambda_ in lambdas:
    fluxes[lambda_] = {}
    
for lambda_ in lambdas:
    for i in [3,4,5]:
            p = anal[i].GaussFitTot(lambda_, 0, remove_max=0, remove_mean=False, plot=False)
            fluxes[lambda_][i] = p.amplitude.value
    
```

```python
err_fluxes = {}

for lambda_ in lambdas:
    err_fluxes[lambda_] = {}
    
for lambda_ in lambdas:
    for i in [3,4,5]:
            err_fluxes[lambda_][i] = anal[i].GetTotBootstrapErrsFit2D(lambda_, 0, remove_mean=False, remove_max=0, nsim=100, nboot=2.)
    
```

```python
err_fluxes
```

```python
plt.figure(figsize=(6,6))

markers = {3:'o', 4:'s', 5:'*'}
colors  = {250:'tomato', 350: 'royalblue', 500:'green'}

plt.title(r'$1 < z < 5$', size=15)

for i in [3,4,5]:
    plt.errorbar(250, -3, fmt=markers[i], color='k', label=r'S/N @ W4 = %d $\quad N_{\rm QSO}=%d $'%(i,len(RA[i])))

for lambda_ in lambdas:
    for i in [3,4,5]:
        plt.errorbar(lambda_+10*i-40, fluxes[lambda_][i], yerr=err_fluxes[lambda_][i], fmt=markers[i], color=colors[lambda_])
    
plt.xlabel(r'$\lambda [\mu$m]', size=15)
plt.ylabel(r'$S_{\nu}$ [mJy]', size=15)
plt.ylim([3,24])
plt.legend(loc='best')

plt.savefig('/Users/fbianchini/Research/StacktSZ/SPIRE_fluxes_W4_selected_SNR_3_4_5.pdf', bbox_inches='tight')
```
